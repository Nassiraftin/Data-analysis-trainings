# Task
Analyze the provided population dataset by continent, including calculating the number of countries per continent, the total population trends over time (for years 2022, 2020, 2015, 2010, 2000, 1990, 1980, 1970), and the average country population per continent for the same years. Visualize these findings with appropriate plots such as bar charts for distribution and current populations, and line plots for population trends.

## Analyze continent distribution

### Subtask:
Get the value counts of the 'Continent' column to understand the number of countries per continent.


**Reasoning**:
To understand the distribution of countries across continents, I will calculate the value counts of the 'Continent' column in the DataFrame.



```python
continent_counts = df['Continent'].value_counts()
print(continent_counts)
```

## Calculate total population trends per continent

### Subtask:
Group the DataFrame by 'Continent' and calculate the sum of populations for each available year (2022, 2020, 2015, 2010, 2000, 1990, 1980, 1970) to observe population trends over time for each continent.


**Reasoning**:
To analyze population trends by continent, I will first define a list of population columns, then group the DataFrame by 'Continent' and calculate the sum of these population columns for each continent.



```python
population_columns = ['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population', '1990 Population', '1980 Population', '1970 Population']
continent_population_trends = df.groupby('Continent')[population_columns].sum()
print(continent_population_trends)
```

**Reasoning**:
To understand the average population per country within each continent, I will divide the total population of each continent (calculated in `continent_population_trends`) by the number of countries in that continent (from `continent_counts`).



```python
average_continent_population = continent_population_trends.div(continent_counts, axis=0)
print(average_continent_population)
```

**Reasoning**:
To visualize the distribution of countries across different continents, I will create a bar plot using the `continent_counts` Series. This will provide a clear representation of how many countries belong to each continent.



```python
plt.figure(figsize=(10, 6))
sns.barplot(x=continent_counts.index, y=continent_counts.values, palette='viridis')
plt.title('Number of Countries per Continent')
plt.xlabel('Continent')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous code produced a `FutureWarning` due to the `palette` argument being used without `hue` in `sns.barplot`. To resolve this, I will explicitly set the `hue` argument to `continent_counts.index` and `legend=False` as suggested by the warning.



```python
plt.figure(figsize=(10, 6))
sns.barplot(x=continent_counts.index, y=continent_counts.values, hue=continent_counts.index, palette='viridis', legend=False)
plt.title('Number of Countries per Continent')
plt.xlabel('Continent')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Reasoning**:
To visualize the total population trends for each continent over the years, I will create a line plot using the `continent_population_trends` DataFrame. This will clearly show how the population of each continent has changed over time.



```python
plt.figure(figsize=(12, 7))
continent_population_trends.T.plot(kind='line', marker='o', figsize=(12, 7))
plt.title('Total Population Trends per Continent (1970-2022)')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Continent', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**Reasoning**:
To visualize the average population per country for each continent over the years, I will create a line plot using the `average_continent_population` DataFrame. This will allow for comparison of average country population trends across continents.



```python
plt.figure(figsize=(12, 7))
average_continent_population.T.plot(kind='line', marker='o', figsize=(12, 7))
plt.title('Average Country Population Trends per Continent (1970-2022)')
plt.xlabel('Year')
plt.ylabel('Average Population per Country')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Continent', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Task
Generate a line plot showing the population growth of USA, China, India, Russia, and Nigeria from 1970 to 2022 using the `df` DataFrame.

## Filter data for specified countries and years

### Subtask:
Select rows corresponding to USA, China, India, Russia, and Nigeria, and relevant population columns from 1970 to 2022.


**Reasoning**:
To filter the DataFrame for the specified countries and population columns, I will first define the lists of target countries and population columns, then filter the original DataFrame based on these lists.



```python
target_countries = ['United States', 'China', 'India', 'Russia', 'Nigeria']
population_columns = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']

filtered_df = df[df['Country/Territory'].isin(target_countries)]
filtered_population_data = filtered_df[['Country/Territory'] + population_columns]

print(filtered_population_data)
```

**Reasoning**:
To prepare the data for plotting population trends over time, I will first melt the `filtered_population_data` DataFrame. This will transform the population columns (years) into rows, making it easier to plot population against years for each country. The `Country/Territory` column will be used as the identifier, and the population columns will be unpivoted.



```python
population_melted = filtered_population_data.melt(id_vars=['Country/Territory'],
                                                  var_name='Year',
                                                  value_name='Population')

population_melted['Year'] = population_melted['Year'].str.replace(' Population', '').astype(int)

print(population_melted.head())
```

**Reasoning**:
To visualize the population growth of the selected countries over time, I will create a line plot using the melted population data, with 'Year' on the x-axis, 'Population' on the y-axis, and different lines for each 'Country/Territory'.



```python
plt.figure(figsize=(12, 7))
sns.lineplot(data=population_melted, x='Year', y='Population', hue='Country/Territory', marker='o')
plt.title('Population Growth of USA, China, India, Russia, and Nigeria (1970-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Country/Territory', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
filtered_population_data_copy = filtered_population_data.copy()
filtered_population_data_copy['Population_Growth'] = filtered_population_data_copy['2022 Population'] - filtered_population_data_copy['1970 Population']

highest_growth_country = filtered_population_data_copy.loc[filtered_population_data_copy['Population_Growth'].idxmax()]

print(f"The country with the highest population growth from 1970 to 2022 is {highest_growth_country['Country/Territory']} with a growth of {highest_growth_country['Population_Growth']:,} people.")
```
